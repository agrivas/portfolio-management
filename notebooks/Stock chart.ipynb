```python
import plotly.graph_objects as go
import yfinance

tsla = yfinance.Ticker("TSLA")
hist = tsla.history(period='1y')

hist['diff'] = hist['Close'] - hist['Open']
hist.loc[hist['diff'] >= 0, 'color'] = 'green'
hist.loc[hist['diff'] < 0, 'color'] = 'red'
hist.head()

hist.head()
```

```python
fig = go.Figure(data = go.Scatter(x = hist.index, y = hist['Close'], mode = 'lines+markers'))
fig.show()
```

```python
from plotly.subplots import make_subplots

fig2 = make_subplots(specs=[[{'secondary_y': True}]])
fig2.add_trace(go.Candlestick(x = hist.index,
                              open = hist['Open'],
                              high = hist['High'],
                              low = hist['Low'],
                              close = hist['Close'],
                              name = 'Price'), secondary_y = False)
fig2.add_trace(go.Bar(x = hist.index, y = hist['Volume'], marker={'color': hist['color']}, name = 'Volume'), secondary_y = True)
fig2.update_yaxes(range = [0,320], secondary_y=False)
fig2.update_yaxes(range = [0,1000000000], secondary_y=True)
fig2.update_yaxes(visible=False, secondary_y=True)
fig2.update_xaxes(rangebreaks = [
    dict(bounds=['sat','mon']),
    #dict(bounds=[16, 9.5], pattern='hour'), # for hourly chart, hide non-trading hours (24hr format)
    dict(values=["2021-12-25","2022-01-01"]) #hide Xmas and New Year
])

fig2.add_trace(go.Scatter(x = hist.index, y = hist['Close'].rolling(20).mean(), name = '20D-MA'))
fig2.update_layout(title = {'text': 'TSLA'}, height = 750)

fig2.show()
```
